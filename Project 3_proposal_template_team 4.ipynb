The main focus of this project will be in predicting the sale price of a car. There are a few ways we can go about this. The first way is that we will try a model to predict just the numerical value of the sale price. The second way to go about this will be to predict whether or not the car sold over the MMR price. Car's are a very big aspect of any American's life, so the knowledge of how much a car is going to cost is something relevant to many people.

We find this project interesting for the same reason. Car's are a very large part of the culture in the U.S. and it will be interesting to see if there are any surprising factors in what makes a car sell well.

The data we found has several descriptive characteristics along with the price so there is lots of data for us to compare here. And I definitely think that this will be a challenging enough question, because the range of factors is very very broad, and predict specific values can oftentimes be very challenging.
